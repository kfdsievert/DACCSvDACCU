# Panel B: Climate Effect of Mitigation Options

This notebook creates panel (b) showing mitigation strategies.

```python3
import matplotlib.pyplot as plt
import numpy as np

# Create a simple test to verify the structure
fig, ax = plt.subplots(figsize=(14, 10))

# DACCS row with emission types
emission_types = ['Total aviation', 'CO₂', 'Contrails', 'NOx', 'H₂O', 'Soot', 'Sulfate']

# DACCS values (same as baseline)
daccs_gwp100 = [-126, 15, 34, 243, 1003, 1541, 2712]
daccs_gwp20 = [-462, 59, 125, 1319, 3556, 1541, 6139]

y_pos = np.arange(len(emission_types))
bar_height = 0.35

color_daccs = '#4BA3BA'  # Teal

# Plot DACCS GWP100 (left bars)
ax.barh(y_pos - bar_height/2, daccs_gwp100, bar_height, color=color_daccs, alpha=0.8, label='DACCS')

# Plot DACCS GWP20 (right bars)
ax.barh(y_pos + bar_height/2, daccs_gwp20, bar_height, color=color_daccs, alpha=0.8)

# Add dots at 0 (remaining emissions)
for i in range(len(emission_types)):
    ax.plot(0, i - bar_height/2, 'o', color='black', markersize=8, zorder=5)
    ax.plot(0, i + bar_height/2, 'o', color='black', markersize=8, zorder=5)

ax.set_yticks(y_pos)
ax.set_yticklabels(emission_types, fontsize=11)
ax.xaxis.tick_top()
ax.axvline(x=0, color='black', linewidth=2)
ax.set_xlim(-7500, 7500)
ax.set_xticks(np.arange(-7500, 8000, 2500))
ax.grid(axis='x', alpha=0.6, linestyle=':', linewidth=1)
ax.set_axisbelow(True)

ax.set_title('DACCS - Climate effect of mitigation (test)', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/test_panel_b.png', dpi=300, bbox_inches='tight')
plt.show()

print("Panel B test created")
```
